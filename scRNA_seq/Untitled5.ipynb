```python
from mpl_toolkits import mplot3d
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x1 = np.linspace(0, 20, 30)
y1 = np.linspace(0, 20, 30)

X1, Y1 = np.meshgrid(x, y)
Z1 = f(X1, Y1)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, Y1, Z1, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, Y+Y1, Z+Z1, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
