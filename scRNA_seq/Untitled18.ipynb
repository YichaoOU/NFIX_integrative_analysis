```r
merged_obj <- SetIdent(merged_obj, value = merged_obj@meta.data$manual_label)
```

```r
markers = read.table("denovo.marker.csv",sep=",")
top_markers = list()
for (i in c(20)){
    top_markers[[paste("top",i,sep="")]] = markers %>%
        group_by(cluster) %>%
        top_n(n = i, wt = avg_log2FC)
}
library(corrplot)

library(RColorBrewer)
options(repr.plot.width = 10, repr.plot.height = 10)
av.exp <- AverageExpression(merged_obj)$RNA

for (i in c(20)){
    data = top_markers[[paste("top",i,sep="")]]
    cor.exp <- cor(av.exp[data$gene,])
    pdf("correlation.top20.pdf")
    corrplot(cor.exp, method = 'number',is.corr = FALSE,col = rev(brewer.pal(n = 10, name = 'RdBu')),order = 'hclust', addrect = 8)
    dev.off()
}
```
